```python
def Bernoulli(p):  #Implementamos una función que represente la función de probabilidad de un evento Bernoulli
    if(random()<=p):
        return 1
    else:
        return 0
```

```python
def Binomial(m,p):           #Implementamos una función binomial a través de nuestra función Bernoulli
    lista=[Bernoulli(p) for i in srange(m)]
    return sum(lista)
```

```python
def sir(beta,gamma,poblacion):  #Creamos nuestra función S.I.R.
    N=poblacion
    I=[1]  #Inicializamos los infectados en t=0
    S=[poblacion] #Inicializamos la población susceptible con toda la población
    R=[0] #Inicializamos los recuperados en 0
    T=[0] #Inicializamos el tiempo en 0
    t=0
    while(I[t]>0):
        s=(S[t]-Binomial(S[t],1-((1-beta)^(I[t]))))
        i=(I[t]+Binomial(S[t],1-((1-beta)^(I[t])))-Binomial(I[t],gamma))
        r=(R[t]+Binomial(I[t],gamma))
        S.append((s))
        I.append(i)
        R.append((r))
        T.append((t+1))
        t+=1
    H=[]
    J=[]
    O=[]
    for s in srange(len(T)):
        H.append([T[s],S[s]])
        J.append([T[s],R[s]])
        O.append([T[s],I[s]])
    return [H,J,O]
```
